```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as mpl
import seaborn as sns
```

```python
lista_fc = pd.read_excel('/Users/giovanni/Desktop/directory_py3/Quotazioni_Fantacalcio_Ruoli_Mantra.xlsx',header=1)
```

```python
lista_fc.sample(1)
```

```python
lista_fc['Assegnato_a']='svincolato'
lista_fc['Prezzo']=0
lista_fc['Svincolato_da']='svincolato'
lista_fc['Prezzo_svinc']=0
lista_fc['called']=False
lista_fc.rename(columns={'R':'RM','Qt. A':'Quotazione','Nome':'NomeCalc'},inplace=True)
lista_fnc=lista_fc[['Id','RM','NomeCalc','Squadra','Quotazione','Assegnato_a','Prezzo',
                    'Svincolato_da','Prezzo_svinc','called']]
lista_fnc.sample(1)
```

```python
lista_fc['RM'].unique()
```

```python
difensori=lista_fc[lista_fc['RM'].str.contains('D')]
portieri=lista_fc[lista_fc['RM'].str.contains('Por')]
centrocampisti=lista_fc[(lista_fc['RM']!= 'Pc') & (lista_fc['RM']!='A') & (lista_fc['RM']!='Por') 
                       & (~lista_fc['RM'].str.contains('D'))]
attaccanti=lista_fc[(lista_fc['RM']=='A') | (lista_fc['RM']=='Pc')]
```

```python
portieri['RC']='P'
difensori['RC']='D'
centrocampisti['RC']='C'
attaccanti['RC']='A'
```

```python
lista_calc=pd.concat([portieri,difensori,centrocampisti,attaccanti])
```

```python
lista_calc.sample(10)
```

```python
lista_calc=lista_calc[['Id','RC','RM','NomeCalc','Squadra','Quotazione','Assegnato_a','Prezzo',
                      'Svincolato_da','Prezzo_svinc','called']]
```

```python
lista_calc.head(1)
```

```python
lista_calc=lista_calc.sort_values(by=['RC','NomeCalc'],ascending=[False,True])
```

```python
lista_calc.head(5)
```

```python
lista_calc['Ordine']=np.random.rand(len(lista_calc['Id']),1)
```

```python
lista_calc=lista_calc.sort_values(by=['RC','Ordine'],ascending=[False,True])
```

```python
residui=lista_calc.groupby(['Assegnato_a']).sum()
#
res=lista_calc.groupby(['RC','Assegnato_a']).count()
#
residui.reset_index(inplace=True)
res.reset_index(inplace=True)
#
residui['Residuo']=600-residui['Prezzo']-residui['Prezzo_svinc']
residui[['Assegnato_a','Prezzo','Prezzo_svinc','Residuo']]
#
situazione=res.merge(residui,on=['Assegnato_a'])
situazione=situazione[['RC','Assegnato_a','Id_x','Prezzo_y','Prezzo_svinc_y','Residuo']]
situazione.sort_values(by=['RC'],ascending=[False],inplace=True)
situazione.sort_values(by=['Residuo'],ascending=[False],inplace=True)
situazione.rename(columns={'Id_x':'Conta','Prezzo_y':'Costo','Prezzo_svinc_y':'Costo_svinc'},inplace=True)
situazione=situazione.pivot_table(index=['Assegnato_a','Residuo'],columns=['RC'],values='Conta')
situazione.reset_index(inplace=True)
situazione=situazione[['Assegnato_a','Residuo','P','D','C','A']]
situazione.sort_values(by='Residuo',ascending=False,inplace=True)
#
situazione[situazione['Assegnato_a']!='svincolato']
```
